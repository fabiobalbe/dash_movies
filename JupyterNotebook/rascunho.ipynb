```python
import pandas as pd
```

```python
data = pd.read_csv('disney_movies.csv')

data
```

```python
data['release_date'] = pd.to_datetime(data['release_date'])

data['Year'] = pd.to_datetime(data['release_date']).dt.year

data
```

```python
data1 = data[data['Year'] >= 2000]

data1.to_csv('update.csv', index=False)
```

```python
df = pd.read_csv('update.csv')

df
```

```python
df['Year'].max()
```

```python
df['genre'].unique()
```

```python
df1 = df.groupby(['genre', 'Year'])['total_gross'].sum().reset_index()

df1
```

```python
df2 = df1[(df1['genre'] == 'Musical') & (df1['Year'] == 2000)]['total_gross'].sum()
df2
```

```python
df3 = df1['total_gross'].sum()
df3
```

```python
df4 = (df2 / df3) * 100
df4
```

```python
g = df[(df['mpaa_rating'] == 'G') & (df['Year'] == 2001)]['mpaa_rating'].count()
p_g = df[(df['mpaa_rating'] == 'PG') & (df['Year'] == 2001)]['mpaa_rating'].count()
p_g13= df[(df['mpaa_rating'] == 'PG-13') & (df['Year'] == 2001)]['mpaa_rating'].count()
r= df[(df['mpaa_rating'] == 'R') & (df['Year'] == 2001)]['mpaa_rating'].count()
nr= df[(df['mpaa_rating'] == 'Not Rated') & (df['Year'] == 2001)]['mpaa_rating'].count()

print(g, p_g, p_g13, r, nr)
```

```python
total_rated = g + p_g + p_g13 + r + nr
total_rated
```

```python
g_per = (g / total_rated) * 100
g_per
```
